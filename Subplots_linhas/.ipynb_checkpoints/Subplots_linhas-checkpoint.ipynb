Schwabish, J. (2014). An economistâ€™s guide to visualizing data. Washington, DC: Urban Institute. Recuperado de https://surf.econ.uic.edu/wp-content/uploads/sites/882/2023/05/Schwabish-Jonathan.-An-Economists-Guide-to-Visualizing-Data-2014.pdf

```python
# -*- coding: utf-8 -*-
# Subplots (2x2) em estilo "paper" com:
# - alinhamento de tÃ­tulos (subplots: left/center; suptitle: left/center relativo aos subplots)
# - controle centralizado de fontes e cores
# - distÃ¢ncia vertical do suptitle parametrizÃ¡vel
# - largura das linhas parametrizÃ¡vel (apenas a destacada mais grossa)

import numpy as np
import matplotlib.pyplot as plt

# =========================== CONFIGS EDITÃVEIS ===========================
STYLE = {
    # Alinhamento dos tÃ­tulos dos subplots: "left" ou "center"
    "subplot_title_loc": "left",

    # Alinhamento do suptitle relativo Ã€ ÃREA DOS SUBPLOTS:
    # "left"  -> x = borda esquerda real dos subplots
    # "center"-> x = centro geomÃ©trico entre bordas esquerda/direita dos subplots
    "suptitle_align": "left",

    # DistÃ¢ncia vertical do suptitle (em coordenadas da figura)
    # Ã‰ somada ao topo real dos subplots: y = top + suptitle_gap
    # Use valores ~0.005 a 0.05 (negativo aproxima ainda mais; positivo afasta).
    "suptitle_gap": 0.05,

    # Tamanhos de fonte (pt)
    "font.size": 10,         # base global
    "axes.titlesize": 12,    # tÃ­tulo de cada subplot
    "axes.labelsize": 11,    # rÃ³tulos de eixos (se usados)
    "xtick.labelsize": 10,   # rÃ³tulos do eixo X
    "ytick.labelsize": 10,   # rÃ³tulos do eixo Y
    "figure.titlesize": 14,  # suptitle

    # EstÃ©tica dos eixos
    "axes.edgecolor": "0.25",
    "axes.linewidth": 1.0,
    "axes.titlepad": 8,      # distÃ¢ncia do tÃ­tulo do subplot para o eixo
    "tick.size": 4,          # tamanho dos ticks

    # Margens (fraÃ§Ãµes da figura) â€” ajuste se precisar mais/menos espaÃ§o externo
    "bottom_margin": 0.06,        # margem inferior p/ rÃ³tulos comuns do eixo X
    "top_margin_for_axes": 0.92,  # topo reservado para eixos (suptitle ficarÃ¡ acima)
}

# Paleta (tons de cinza). Troque aqui se quiser colorir.
COLORS = {
    "highlight": "0.05",   # linha destacada (quase preto)
    "normal": "0.45",      # linhas normais (cinza mÃ©dio)
    "grid_major": "0.70",  # linhas horizontais de referÃªncia
    "grid_minor": "0.90",  # linhas verticais (anos)
    "text": "0.05",        # cor do texto (suptitle, etc.)
}

# Larguras de linha (apenas a destacada mais grossa)
LW = {
    "highlight": 3.0,   # espessura da sÃ©rie destacada
    "normal": 1.6,      # espessura das sÃ©ries nÃ£o-destaque
    "grid_major": 0.5,  # espessura das linhas horizontais
    "grid_minor": 0.5,  # espessura das linhas verticais
}

# =========================== APLICA O ESTILO GLOBAL (rcParams) ===========================
plt.rcParams.update({
    "font.family": "serif",
    "font.serif": ["DejaVu Serif", "Times New Roman", "Times"],

    # AparÃªncia dos eixos
    "axes.edgecolor": STYLE["axes.edgecolor"],
    "axes.linewidth": STYLE["axes.linewidth"],
    "axes.titlepad": STYLE["axes.titlepad"],

    # Ticks
    "xtick.direction": "out",
    "ytick.direction": "out",
    "xtick.major.size": STYLE["tick.size"],
    "ytick.major.size": STYLE["tick.size"],

    # Fontes/tamanhos
    "font.size": STYLE["font.size"],
    "axes.titlesize": STYLE["axes.titlesize"],
    "axes.labelsize": STYLE["axes.labelsize"],
    "xtick.labelsize": STYLE["xtick.labelsize"],
    "ytick.labelsize": STYLE["ytick.labelsize"],
    "figure.titlesize": STYLE["figure.titlesize"],
})

# =========================== DADOS SIMULADOS ===========================
np.random.seed(42)
years = np.arange(0, 6, 1)  # 0..5 (seis pontos)

def shape_trend(base_points, noise=0.6):
    """
    Gera sÃ©rie com ruÃ­do ao redor de 'base_points'.
    base_points: lista de 6 nÃºmeros (um por ano).
    noise: desvio padrÃ£o do ruÃ­do Gaussiano.
    """
    base = np.array(base_points, dtype=float)
    return base + np.random.normal(0, noise, size=base.shape)

# TrÃªs sÃ©ries por painel (%)
panels = {
    "Adult Only": np.vstack([
        shape_trend([4, 5, 9, 9,10,12]),
        shape_trend([4, 5, 6, 6, 7, 9]),
        shape_trend([5, 5, 6, 7, 9,10]),
    ]),
    "No Cash": np.vstack([
        shape_trend([1, 3,11, 9, 9,10]),
        shape_trend([0.5, 2, 5,3.5,5.1,6.2]),
        shape_trend([0, 2, 1,-3,-2,-1]),
    ]),
    "Welfare Cash Assistance": np.vstack([
        shape_trend([2, 0, 2, 5, 4, 6]),
        shape_trend([5, 6.5, 8,12.5,12.8,18]),
        shape_trend([3, 3, 6, 9, 7.6,12.4]),
    ]),
    "SSI Cash Assistance": np.vstack([
        shape_trend([0, 2,-3, 0,-6.5,-7.5]),
        shape_trend([5, 8, 5.5, 5.5, 3.5, 4]),
        shape_trend([-5,-1.5,-10.5,-7,-15.2,-16]),
    ]),
}

# SÃ©rie destacada (0, 1 ou 2) por painel
highlight_choices = {
    "Adult Only": 2,
    "No Cash": 0,
    "Welfare Cash Assistance": 2,
    "SSI Cash Assistance": 1,
}

# =========================== HELPERS DE LAYOUT ===========================
def _axes_bounds(axes_collection):
    """
    Retorna (left, right, top, bottom) das BBox dos subplots em coords da figura (0..1).
    Aceita lista/tupla/ndarray de Axes.
    """
    axs = np.ravel(axes_collection)
    lefts   = [ax.get_position().x0 for ax in axs]
    rights  = [ax.get_position().x1 for ax in axs]
    tops    = [ax.get_position().y1 for ax in axs]
    bottoms = [ax.get_position().y0 for ax in axs]
    return min(lefts), max(rights), max(tops), min(bottoms)

def place_suptitle(fig, axes_collection, text, align="left",
                   fontsize=None, color=COLORS["text"], gap=0.02):
    """
    Coloca o suptitle alinhado Ã€ ÃREA DOS SUBPLOTS, com distÃ¢ncia vertical parametrizÃ¡vel.
    - align: "left"  -> x = left (borda esquerda dos subplots)
             "center"-> x = (left + right)/2
    - gap: distÃ¢ncia vertical (em coords da figura) adicionada ao topo dos subplots.
           Ex.: 0.02 â‰ˆ 2% da altura da figura acima do topo.
           Pode ser negativo para aproximar mais.
    """
    if fontsize is None:
        fontsize = plt.rcParams["figure.titlesize"]

    # Garante que as posiÃ§Ãµes dos eixos estejam atualizadas
    fig.canvas.draw()

    # Mede bordas reais dos subplots
    left, right, top, _ = _axes_bounds(axes_collection)

    # PosiÃ§Ã£o horizontal conforme alinhamento desejado
    if align == "left":
        x = left; ha = "left"
    else:  # "center"
        x = 0.5 * (left + right); ha = "center"

    # PosiÃ§Ã£o vertical: topo dos eixos + gap (com clamp para nÃ£o ultrapassar o canvas)
    y = max(0.0, min(0.995, top + gap))

    # Insere o suptitle como fig.text para controle fino de x/y
    fig.text(x, y, text, ha=ha, va="bottom", fontsize=fontsize, color=color)

# =========================== DESENHO DOS SUBPLOTS ===========================
def draw_panel(ax, title, data, highlight_idx,
               colors=COLORS, lw=LW, title_loc="left"):
    """
    Desenha um painel:
    - linhas horizontais principais (-20, -10, 0, 10, 20)
    - linhas verticais fracas nos anos internos (1..4)
    - 3 sÃ©ries (apenas a destacada com espessura maior)
    - tÃ­tulo alinhado via title_loc
    """
    # Linhas horizontais (grid principal)
    for y in [-20, -10, 0, 10, 20]:
        ax.axhline(y, color=colors["grid_major"], linewidth=lw["grid_major"], zorder=0)

    # Linhas verticais suaves (marcam anos intermediÃ¡rios)
    for x in years[1:-1]:
        ax.axvline(x, color=colors["grid_minor"], linewidth=lw["grid_minor"],
                   linestyle="--", zorder=0)

    # SÃ©ries: uma destacada (largura 'highlight') e demais com 'normal'
    for i in range(data.shape[0]):
        width = lw["highlight"] if i == highlight_idx else lw["normal"]
        color = colors["highlight"] if i == highlight_idx else colors["normal"]
        ax.plot(years, data[i], color=color, linewidth=width, zorder=2+i)

    # Limites e ticks
    ax.set_xlim(0, 5)
    ax.set_ylim(-20, 20)
    ax.set_xticks(years)
    ax.set_yticks(np.arange(-20, 21, 10))

    # TÃ­tulo do subplot (alinhamento controlado)
    ax.set_title(title, loc=title_loc, color=COLORS["text"])

    # Estilo mais "clean"
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)

# =========================== FIGURA (2x2) ===========================
fig, axes = plt.subplots(2, 2, figsize=(11, 8), sharex=True, sharey=True)
ax_list = axes.ravel()
titles = list(panels.keys())

# Desenha cada painel
for ax, title in zip(ax_list, titles):
    draw_panel(
        ax, title, panels[title],
        highlight_choices.get(title, 0),
        colors=COLORS,
        lw=LW,
        title_loc=STYLE["subplot_title_loc"]
    )

# 1) Layout preliminar para estabilizar posiÃ§Ãµes
fig.tight_layout()
fig.canvas.draw()

# 2) Mede bordas reais e reaplica tight_layout com retÃ¢ngulo baseado nessas bordas
left, right, top, bottom = _axes_bounds(ax_list)
fig.tight_layout(rect=[left, STYLE["bottom_margin"], right, STYLE["top_margin_for_axes"]])
fig.canvas.draw()

# 3) SUPTITLE: alinhamento relativo aos subplots + distÃ¢ncia vertical parametrizÃ¡vel
place_suptitle(
    fig, ax_list,
    "Implied Impulse Response Functions for Different Caseloads\n(Percent change)",
    align=STYLE["suptitle_align"],
    fontsize=STYLE["figure.titlesize"],
    color=COLORS["text"],
    gap=STYLE["suptitle_gap"]     # <<< ajuste aqui a distÃ¢ncia vertical
)

# 4) RÃ³tulos comuns de eixos (opcional)
fig.text(0.5, 0.04, "Years", ha="center", color=COLORS["text"])
fig.text(0.06, 0.5, "Percent change", va="center", rotation="vertical", color=COLORS["text"])

# Salvar/mostrar
fig.savefig("linhas_subplots.png", dpi=300, bbox_inches="tight")
plt.show()
```
