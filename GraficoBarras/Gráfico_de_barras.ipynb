```python
#Importa as bibliotecas
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.lines import Line2D
```

```python
# Dados
regions = ['East', 'South', 'North', 'West']
furniture = [39, 32, 23, 18]
office_supplies = [45, 35, 27, 21]
software = [37, 29, 23, 17]
total_revenue = [121, 97, 73, 55]
```

```python
# Informações para os gráficos
x = np.arange(len(regions))
width = 0.25
bg_width = 0.8

# Cores
colors = {
    'Furniture': '#2f4f4f',
    'Office Supplies': '#aec6cf',
    'Software': '#5f9ea0',
    'Total': 'lightgray'
}

# Criar o gráfico
fig, ax = plt.subplots(figsize=(5, 5))
#fig, ax = plt.subplots(figsize=(10, 6)) # melhor tamanho de gráfico

# Barra de fundo (total)
ax.bar(x, total_revenue, width=bg_width, color=colors['Total'], zorder=0)

# Barras por categoria
ax.bar(x - width, furniture, width, color=colors['Furniture'], zorder=3)
ax.bar(x, office_supplies, width, color=colors['Office Supplies'], zorder=3)
ax.bar(x + width, software, width, color=colors['Software'], zorder=3)

# Rótulos dentro da barra cinza, em azul escuro
for i in range(len(x)):
    ax.text(x[i], total_revenue[i] - 5, f'{total_revenue[i]}M',
            ha='center', va='top', fontsize=10, color='#2f4f4f', fontweight='bold')

# Estética dos eixos
ax.set_ylabel('Revenue')
ax.set_xlabel('Region')
ax.set_title('Revenue by Region and Product Category', loc='left', fontsize=14, weight='bold')
ax.set_xticks(x)
ax.set_xticklabels(regions)

# Remove bordas da área de plotagem, exceto eixo x e y
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('gray')
ax.spines['bottom'].set_color('gray')

# Legenda com círculos e alinhada à direita, espaçamento reduzido
legend_elements = [
    Line2D([0], [0], marker='o', color='w', label='Furniture', markerfacecolor=colors['Furniture'], markersize=10),
    Line2D([0], [0], marker='o', color='w', label='Office Supplies', markerfacecolor=colors['Office Supplies'], markersize=10),
    Line2D([0], [0], marker='o', color='w', label='Software', markerfacecolor=colors['Software'], markersize=10),
]

ax.legend(
    handles=legend_elements,
    loc='upper right',
    bbox_to_anchor=(1, 1.02),
    frameon=False,
    ncol=3,
    handletextpad=0.4,
    columnspacing=0.8
)

# Layout
plt.tight_layout()

# Salva o gráfico em imagem
plt.savefig('barras.png', dpi=300, bbox_inches='tight')
plt.show()
```
