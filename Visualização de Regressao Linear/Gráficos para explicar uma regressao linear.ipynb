### 9 gráficos para explicar a regressão linear

```python
# Importacao das bibliotecas
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
import shap
```

```python
#  Carregamento e preparação dos dados

# Carrega o dataset de score de crédito
credit_score = pd.read_csv("credit_score.csv")  # Ajuste o caminho conforme necessário

# Variável alvo (score de crédito)
y = credit_score['CREDIT_SCORE']

# Cria variável com quadrado da razão de gastos
credit_score['R_EXPENDITURE_SQR'] = credit_score['R_EXPENDITURE'] ** 2

# Seleciona as colunas desejadas diretamente do DataFrame original (sem gerar cópia ambígua)
X = credit_score[['INCOME', 'R_DEBT_INCOME', 'R_EXPENDITURE', 'R_EXPENDITURE_SQR', 'CAT_GAMBLING']].copy()

# Cria as variáveis dummies de forma segura com .loc
X.loc[:, 'GAMBLING_LOW'] = X['CAT_GAMBLING'].apply(lambda x: 1 if x == 'Low' else 0)
X.loc[:, 'GAMBLING_HIGH'] = X['CAT_GAMBLING'].apply(lambda x: 1 if x == 'High' else 0)

# Remove a coluna original
X.drop(columns=['CAT_GAMBLING'], inplace=True)
```

```python
credit_score.head()
```

```python
# Ajuste do modelo de regressão linear

# Adiciona uma constante (intercepto) ao modelo
X = sm.add_constant(X)

# Ajusta o modelo OLS (Mínimos Quadrados Ordinários)
model = sm.OLS(y, X).fit()

# Exibe o resumo do modelo
print(model.summary())
```

#### Plot 1 - Gráfico de resíduos (valores previstos vs reais)

```python
# Obtém os valores previstos
y_pred = model.predict(X)

# Cria o gráfico de dispersão
fig, ax = plt.subplots(figsize=(5, 5))
plt.scatter(y_pred, y)
axes_fonts = {'fontsize': 10}

# Linha vermelha y = x (linha de previsão perfeita)
ax.plot([y.min(), y.max()], [y.min(), y.max()], color='tab:red')

# Rótulos dos eixos
plt.xlabel('Previsto', size=7)
plt.ylabel('Real', size=7)


# Define o título do gráfico com formatação
ax.set_title(
    'Gráfico de dispersao',
    weight='bold', size=12, pad=20
)

# Salva o gráfico em imagem
plt.savefig('grafico_dispersao.png', dpi=300, bbox_inches='tight')


plt.show()
```

#### 📊 Interpretação dos Resultados do Gráfico de Dispersão

O gráfico gerado mostra a relação entre os valores previstos pelo modelo de regressão linear (y_pred) e os valores reais (y) da variável de interesse (por exemplo, score de crédito).

##### ✅ O que foi observado:
A maioria dos pontos está relativamente próxima da linha vermelha y = x, o que indica que o modelo foi capaz de aproximar-se dos valores reais com boa precisão.

Existe uma variação natural em torno da linha, o que é esperado em modelos preditivos aplicados a dados reais.

Não há evidência clara de viés sistemático (por exemplo, não parece que o modelo superestima ou subestima consistentemente em alguma faixa).

Os pontos não seguem um padrão curvo ou funil, o que sugere que a variância dos erros é razoavelmente constante (homocedasticidade).

Não foram detectadas concentrações anômalas de pontos em regiões específicas, o que reforça a robustez geral do modelo nesse conjunto de dados.

##### ⚠️ Possíveis limitações visíveis:
Ainda que os pontos estejam bem distribuídos, existem alguns casos mais distantes da linha — essas são previsões com erro maior (outliers do modelo).

Isso pode ser normal, mas vale a pena analisar esses casos separadamente: talvez o modelo não consiga explicar bem alguns perfis de clientes.

###### 📌 Conclusão:
O gráfico sugere que:

O modelo está fazendo boas previsões em geral

Erros estão razoavelmente distribuídos

Não há viés evidente ou padrão problemático nos resíduos

Para uma análise mais completa, é recomendável gerar também:

Gráfico de resíduos (y - y_pred)

Histograma dos erros

Gráfico de resíduos vs. variáveis explicativas

Essas análises complementam a visualização e ajudam a identificar eventuais falhas estruturais no modelo.



#### Plot 2 - Heatmap de correlação entre as variáveis

```python
# Matriz de correlação sem a constante
corr_matrix = X.drop('const', axis=1).corr()
fig, ax = plt.subplots(figsize=(5, 5))
axes_fonts = {'fontsize': 7}
# Mapa de calor das correlações
sns.heatmap(corr_matrix, cmap="coolwarm", fmt=".1f", vmin=-1, vmax=1, annot=True)
plt.title('Matriz de Correlação das Variáveis')

# Salva o gráfico em imagem
plt.savefig('matriz_correlacao.png', dpi=300, bbox_inches='tight')
plt.show()
```

##### 🧠 Interpretação da Matriz de Correlação:
**Correlação +1.0:** Relação linear perfeita e positiva

**Correlação -1.0:** Relação linear perfeita e negativa

**Correlação 0:** Sem relação linear

Essa análise é especialmente útil para:

Detectar multicolinearidade: duas variáveis com correlação próxima de 1 ou -1 podem causar problemas na regressão (inflar variância dos coeficientes).

Verificar redundância: se duas variáveis têm correlação muito alta, pode ser melhor remover uma delas.

Explorar relações entre variáveis: entender se há tendência natural entre, por exemplo, INCOME e R_DEBT_INCOME.

#### Plot 3 - Coeficientes do modelo com erro padrão

```python
# Obtém coeficientes (pesos) e erros padrão, excluindo o intercepto
coefficients = model.params[1:][::-1]
se = model.bse[1:][::-1]
features = X.columns[1:][::-1]

plt.figure(figsize=(5, 5))

# Linha vertical no zero para referência
plt.axvline(x=0, color='tab:red', linestyle='--')

# Gráfico de erro padrão com coeficientes
plt.errorbar(coefficients, features, xerr=se, fmt='o', capsize=5)
plt.xlabel('Coeficiente (Peso)', size=7)
plt.title('Pesos dos Atributos e Incerteza')
axes_fonts = {'fontsize': 10}

# Salva o gráfico em imagem
plt.savefig('incerteza.png', dpi=300, bbox_inches='tight')

plt.show()
```

#### 🧠 Interpretação do gráfico:
Cada ponto azul representa um coeficiente estimado para uma variável.

As linhas horizontais (barras de erro) representam a incerteza associada à estimativa (erro padrão).

A linha vermelha vertical no zero serve como referência:

Se a barra cruza o zero, o coeficiente pode não ser estatisticamente significativo.

Se a barra não cruza o zero, temos mais confiança na direção do efeito (positivo ou negativo).

#### ✅ Conclusão:
Este gráfico é útil para:

Identificar quais variáveis têm maior ou menor impacto na variável resposta.

Avaliar a significância dos coeficientes com base na sua incerteza.

Apoiar uma explicação mais transparente e visual do modelo.



####  Plot 4 – Gráfico de Efeitos Individuais (Effect Plot com boxplot)

```python
# Calcula o efeito de cada variável (valor da variável × coeficiente)
feature_effects = X * model.params

# Combina os efeitos das variáveis derivadas para facilitar a leitura
feature_effects['R_EXPENDITURE'] = feature_effects['R_EXPENDITURE'] + feature_effects['R_EXPENDITURE_SQR']
feature_effects['GAMBLING'] = feature_effects['GAMBLING_LOW'] + feature_effects['GAMBLING_HIGH']

# Remove colunas desnecessárias (constante e variáveis intermediárias)
feature_effects.drop(['const', 'R_EXPENDITURE_SQR', 'GAMBLING_LOW', 'GAMBLING_HIGH'], axis=1, inplace=True)

# Cria o boxplot dos efeitos
plt.figure(figsize=(5, 5))
plt.axvline(x=0, color='tab:red', linestyle='--')  # linha no zero como referência
sns.boxplot(data=feature_effects, orient="h", color="tab:blue")
plt.xlabel("Efeito no Score de Crédito", size=7)
plt.title("Distribuição dos Efeitos das Variáveis")
axes_fonts = {'fontsize': 10}

# Salva o gráfico em imagem
plt.savefig('boxplot.png', dpi=300, bbox_inches='tight')
plt.show()
```

#### 📊 Interpretação do Gráfico:
Cada boxplot horizontal representa a distribuição dos efeitos individuais de uma variável para todos os clientes.

A linha vermelha vertical no x = 0 ajuda a visualizar se os efeitos são positivos, negativos ou mistos.

Os efeitos são calculados para cada indivíduo como:

Efeito = Valor da variável * Coeficiente do modelo

**Distribuições mais à direita** indicam que aquela variável geralmente aumenta o score.

**Distribuições à esquerda** indicam que a variável tende a reduzir o score.

**Distribuições simétricas** em torno de 0 indicam que o efeito varia muito entre os indivíduos.

#### ✅ Conclusão:
Este gráfico é útil para:

Entender o **efeito real e individualizado** de cada variável no modelo.

Comparar **qual variável tem maior ou menor impacto no score** dos clientes.

Comunicar os resultados de forma intuitiva, mesmo para quem não conhece regressão linear.

#### Plot 5 – Efeito Ajustado com Intercepto e Sem Outliers

```python
# Recalcula os efeitos para adicionar o intercepto em todos os casos
feature_effects = X * model.params
const = feature_effects['const'][0]

# Combina os efeitos novamente
feature_effects['R_EXPENDITURE'] = feature_effects['R_EXPENDITURE'] + feature_effects['R_EXPENDITURE_SQR']
feature_effects['GAMBLING'] = feature_effects['GAMBLING_LOW'] + feature_effects['GAMBLING_HIGH']

# Remove colunas auxiliares
feature_effects.drop(['const', 'R_EXPENDITURE_SQR', 'GAMBLING_LOW', 'GAMBLING_HIGH'], axis=1, inplace=True)

# Adiciona o valor da constante (intercepto) a todos os efeitos
feature_effects = feature_effects + const

# Cria gráfico ajustado
plt.figure(figsize=(5, 5))
plt.axvline(x=const, color='tab:red', linestyle='--')  # linha no intercepto
sns.boxplot(data=feature_effects, orient="h", color="tab:blue", showfliers=False)  # exclui outliers
plt.xlabel("Efeito no Score de Crédito", size=7)
plt.title("Efeito das Variáveis (com Intercepto e sem Outliers)")

axes_fonts = {'fontsize': 10}
# Salva o gráfico em imagem
plt.savefig('boxplot_com_intercepto.png', dpi=300, bbox_inches='tight')
plt.show()
```

#### 📊 Interpretação do gráfico:

* Cada boxplot mostra a distribuição do efeito total de uma variável (incluindo intercepto) na predição final.

* A linha vermelha vertical indica o valor do intercepto (ou seja, o ponto de partida da previsão).

* O gráfico exclui os outliers para facilitar a leitura e focar na variação central dos efeitos.

#### 🧠 O que o gráfico revela:
* Variáveis com efeito positivo mais concentrado à direita aumentam o score de crédito na maioria dos casos.

* Variáveis com efeito negativo mais à esquerda reduzem o score.

* A largura dos boxplots indica a variabilidade do impacto da variável entre os clientes.

Isso permite identificar quais variáveis têm efeitos mais consistentes (boxplot estreito) e quais variam muito entre os perfis (boxplot largo).

#### Plot 6 – Gráfico de Importância (Efeito Médio Absoluto)

```python
# Recalcula efeitos para média dos valores absolutos
feature_effects = X * model.params
feature_effects['R_EXPENDITURE'] = feature_effects['R_EXPENDITURE'] + feature_effects['R_EXPENDITURE_SQR']
feature_effects['GAMBLING'] = feature_effects['GAMBLING_LOW'] + feature_effects['GAMBLING_HIGH']
feature_effects.drop(['const', 'R_EXPENDITURE_SQR', 'GAMBLING_LOW', 'GAMBLING_HIGH'], axis=1, inplace=True)

# Cálculo dos efeitos absolutos médios
feature_effects = abs(feature_effects)
mean_effects = feature_effects.mean(axis=0)
mean_effects.sort_values(inplace=True)

# Cria o gráfico de barras
plt.figure(figsize=(5, 5))
plt.barh(mean_effects.index, mean_effects)
plt.xlabel('Efeito Médio Absoluto no Score de Crédito', size=7)
plt.title("Importância das Variáveis na Predição")

# Salva o gráfico em imagem
plt.savefig('Barra.png', dpi=300, bbox_inches='tight')
plt.show()
```

#### ✅ Conclusão:
Este gráfico oferece uma visualização mais próxima da realidade do modelo, pois mostra os efeitos reais previstos, já somando a constante. É útil para:

Explicar o funcionamento do modelo de forma intuitiva

Mostrar a contribuição real de cada variável para o resultado final

Apoiar decisões baseadas em impactos esperados por perfil de cliente

#### Plot 7 – Gráfico de Efeitos Individuais por Cliente (local explanation)

```python
# Reutiliza os efeitos ajustados com intercepto
plt.figure(figsize=(5, 5))
plt.axvline(x=const, color='black', linestyle='--')
sns.boxplot(data=feature_effects, orient="h", color="tab:blue", showfliers=False)
plt.xlabel("Efeito no Score de Crédito", size=7)
plt.title("Efeitos Individuais com Destaque para um Cliente")

axes_fonts = {'fontsize': 10}

# Seleciona o índice de um cliente
idx = 0  # pode mudar para outro cliente

# Calcula os efeitos do cliente individualmente
ind_feature_effect = X.iloc[idx] * model.params[1:]
ind_feature_effect['R_EXPENDITURE'] = ind_feature_effect['R_EXPENDITURE'] + ind_feature_effect['R_EXPENDITURE_SQR']
ind_feature_effect['GAMBLING'] = ind_feature_effect['GAMBLING_LOW'] + ind_feature_effect['GAMBLING_HIGH']
ind_feature_effect.drop(['R_EXPENDITURE_SQR', 'GAMBLING_LOW', 'GAMBLING_HIGH', 'const'], inplace=True)

# Soma o intercepto
ind_feature_effect = ind_feature_effect + const

# Plota o X vermelho nos efeitos do cliente
for i, feature in enumerate(ind_feature_effect):
    plt.scatter(feature, i, c='tab:red', marker='x', s=100, zorder=7)

# Salva o gráfico em imagem
plt.savefig('boxplot_efeito_individual.png', dpi=300, bbox_inches='tight')

plt.show()
```

#### Plot 9 – SHAP (Waterfall e Gráfico de Importância)

```python
import shap
import matplotlib.pyplot as plt

# 1. Amostragem para background
background = shap.sample(X, 100, random_state=42)

# 2. Explicador com SHAP Kernel
explainer = shap.KernelExplainer(model.predict, background)

# 3. Valores SHAP
shap_values = explainer.shap_values(X)

# 4. Gráfico Waterfall para o primeiro cliente (em tamanho reduzido)
explanation = shap.Explanation(
    values=shap_values[0],
    base_values=explainer.expected_value,
    data=X.iloc[0]
)

# Criar figura manualmente para tentar controlar o tamanho
fig = plt.figure(figsize=(5, 5))
shap.plots.waterfall(explanation, show=False)
plt.savefig("shap_waterfall_cliente_0.png", dpi=300, bbox_inches="tight")
plt.show()

# 5. Média prevista
preds = model.predict(X)
print("Score médio previsto:", np.mean(preds))

# 6. Gráfico de importância global com tamanho reduzido
fig = plt.figure(figsize=(5, 5))
shap.summary_plot(shap_values, X, plot_type="bar", show=False)
plt.savefig("shap_importancia_global.png", dpi=300, bbox_inches="tight")
plt.show()
```

#### 🔹 1. Gráfico Waterfall (cliente 0)
Esse gráfico mostra como o modelo chegou à previsão para um cliente específico (no caso, X.iloc[0]). Aqui está como interpretá-lo:

✅ Componentes do gráfico waterfall:
Base Value: é o valor médio da previsão no conjunto de treinamento (background). É o ponto de partida do modelo.

Features: cada linha representa uma variável (feature) e o quanto ela empurrou a previsão para cima ou para baixo.

Setas azuis/vermelhas:

🔴 Vermelho: aumentaram a previsão (efeito positivo)

🔵 Azul: reduziram a previsão (efeito negativo)

Output value (no topo): o valor final da previsão para o cliente 0, após considerar os efeitos de todas as variáveis.

📌 Exemplo de leitura:
A previsão média do modelo era 0.45. A variável idade contribuiu com -0.10, renda com +0.15, tempo_de_cadastro com -0.03, etc. A soma dessas contribuições levou a uma previsão final de 0.47 para esse cliente.

#### 🔹 2. Gráfico de barras (Importância Global)
Esse gráfico mostra a importância média de cada variável para o modelo como um todo, calculada a partir dos valores absolutos dos SHAP values.

✅ Como interpretar:
Cada barra representa uma feature.

O tamanho da barra mostra o impacto médio dessa feature na saída do modelo — ou seja, quanto ela costuma mudar a previsão.

Não indica se a variável aumenta ou reduz a previsão, só o quão relevante ela é.

#### 📌 Exemplo de leitura:
A variável renda é a mais importante para o modelo, seguida por idade, número de dependentes, etc. Essas são as variáveis que mais impactam as decisões do modelo.

#### 📊 Conclusão geral:
O waterfall é explicação individual: “por que o modelo previu esse valor para esse cliente?”

O summary bar é explicação global: “quais variáveis o modelo mais usa para tomar decisões?”
